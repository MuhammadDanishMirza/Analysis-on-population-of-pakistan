# Pakistan Population Census 2017 Dataset

## **About the Dataset**

This dataset contains demographic information from the **Pakistan Population Census** conducted in 2017. It provides detailed population data at various administrative levels, including **provinces**, **divisions**, **districts**, and **sub-divisions**. 

Years of Data:
- 1997
- 2017

The dataset covers:
- Urban and rural populations
- Gender distribution
- Transgender individuals
- Sex ratios
- Population figures from the **1998 census**
- Annual growth rates

---

## **Features**

- **Province**: The administrative provinces or regions where the census data was collected.
- **Division**: The divisions within each province (second-level administrative divisions).
- **District**: Larger administrative units within each division.
- **Sub-Division**: Localized administrative units (tehsils) within each district.
- **Area**: The land area (in square kilometers) of each sub-division.
- **Urban Population 2017**: The population of urban areas in each sub-division for 2017.
- **Rural Population 2017**: The population of rural areas in each sub-division for 2017.
- **Male Population 2017**: The male population within each sub-division for 2017.
- **Female Population 2017**: The female population within each sub-division for 2017.
- **Transgender Population 2017**: The transgender population within each sub-division for 2017.
- **Sex Ratio 2017**: The sex ratio, calculated as the number of females per 1000 males in 2017.
- **Population in 1998**: The total population recorded in each sub-division during the 1998 census.
- **Annual Growth Rate**: The percentage increase in population from 1998 to 2017 for each sub-division.
